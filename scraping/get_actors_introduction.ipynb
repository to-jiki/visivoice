```python
import csv
import json
import codecs
from bs4 import BeautifulSoup
import requests
import time
import json
import codecs
```

```python
actor_name_list = []
with open('../data/voice_actors_greater_than_100characters.csv') as f:
    reader = csv.reader(f,delimiter=",")
    for row in reader:
        actor_name_list.append(row[0])
    del actor_name_list[0]
    actor_name_list = list(dict.fromkeys(actor_name_list))
    print(actor_name_list)
```

```python
actor_number_starting_scraping = 0

top_char_b =["a","i","u","e","o"]
top_char_s =["", "k","s","t","n","h","m","y","r","w"]
top_char = []
for shiin in top_char_s:
    for boin in top_char_b:    
        if((shiin == "y" and (boin == "i" or boin == "e")) or ((shiin =="w") and boin != "a" )):
            continue
        top_char.append(shiin + boin)
        
url_names = []
for link in  top_char:
    url_names.append("https://sakuhindb.com/anime/alph_info/j/people_voice_" +link+".html")

actor_urls = []
all_actor_number = 0

for urlName in url_names:
    url = requests.get(urlName)
    soup = BeautifulSoup(url.content, "html.parser")
    
    all_man = soup.find_all(attrs = {"class":["man","female","no_sex"]})
    for actor in all_man:
        tag = actor.parent['href']
        name = actor.contents[0]
        if(name in actor_name_list):
            actor_urls.append([name, "https://sakuhindb.com/" +tag])
            
            all_actor_number += 1
    print(f"get {all_actor_number} actor's URL")

print(actor_urls)
```

```python
# json形式
# data_json = {}
# csvファイルのpath
filename = '../data/voice_actors_introduction.json'
'''if actor_number_starting_scraping == 0:
    with open(filename, 'a', encoding='UTF-8') as f:
        writer = csv.writer(f)
        writer.writerow(['name', 'jenre', 'title', 'character', 'year'])'''

dict_actors = {}

current_number = 0


# = ["https://sakuhindb.com/tj/6_BFC0B5DCBBFBCCEFC0B8/"]
#otameshi_actor_urls = ["https://sakuhindb.com/tj/6_A5B8A5EBA1A6A5AFA5ECA5A4A5D0A1BCA5B0/"]

for actor_url in actor_urls:
    #print(f"{current_number}/{all_actor_number} done")
    current_number += 1

    #途中で取得失敗したときのためのメニュー
    if current_number < actor_number_starting_scraping:
        continue

    url = requests.get(actor_url[1])
    soup = BeautifulSoup(url.content, "html.parser")
    #print(soup)
    
    name = actor_url[0]
    dict_actors[name] = {}

    info_all = soup.find(attrs = {"class","table width_100per"})
    info_left = info_all.find_all(attrs = {"class","leftmenu"})
    for tag in info_left:
        key = tag.contents[0]
        if(key != "経歴/説明"):
            info_right = tag.next_sibling.next_sibling.next_sibling
            value = info_right.contents[0].string
            if (not (key == "英語名" or key == "職業" or key == "作成者 [" or key == "最終変更者/日時 ")):
                #print(key, value)
                dict_actors[name][key] = value
        else:
            intro_sentense = ""
            info_right = tag.next_sibling.next_sibling.next_sibling

            for tagtag in info_right:
                #print(tagtag.string)
                if(tagtag.string == None):
                    intro_sentense += "<br>"
                else:
                    intro_sentense += tagtag.string

            #print(key,intro_sentense)
            dict_actors[name][key] = intro_sentense

    print(name ,":corrected")
    
    
file_path = '../data/voice_actors_introduction_fixed.json'
outfile = codecs.open(file_path, 'w','utf-8')
json.dump(dict_actors, outfile, ensure_ascii=False)
#print("all process done")
```
